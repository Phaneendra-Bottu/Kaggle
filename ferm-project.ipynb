```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf

train_dir = '/kaggle/input/fer2013-mine/data/train'
val_dir = '/kaggle/input/fer2013-mine/data/test'

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(48,48),
    batch_size=128,  # Increased batch size
    color_mode="grayscale",
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(48,48),
    batch_size=128,  # Increased batch size
    color_mode="grayscale",
    class_mode='categorical'
)

# Learning rate increased slightly to speed up initial learning
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.001,  # Increased initial learning rate
    decay_steps=1 / 1e-6,
    decay_rate=1.0,
    staircase=True)

emotion_model = Sequential()
emotion_model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(48,48,1)))
emotion_model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2,2)))
emotion_model.add(Dropout(0.35))  # Increased dropout rate

emotion_model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2,2)))
emotion_model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2,2)))
emotion_model.add(Dropout(0.4))  # Increased dropout rate

emotion_model.add(Flatten())
emotion_model.add(Dense(1024, activation='relu'))
emotion_model.add(Dropout(0.55))  # Increased dropout rate
emotion_model.add(Dense(7, activation='softmax'))

emotion_model.compile(loss='categorical_crossentropy',
                      optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule),
                      metrics=['accuracy'])

print(emotion_model.summary())

emotion_model_info = emotion_model.fit(
    train_generator,
    steps_per_epoch=28709 // 128,  # Adjusted for new batch size
    epochs=150,  # Increased epochs for better learning
    validation_data=val_generator,
    validation_steps=7178 // 128  # Adjusted for new batch size
)

emotion_model.save_weights('model_improved.weights.h5')
```

```python
import numpy as np

def smooth_curve(y, factor=0.9):
    smoothed = []
    for i in range(len(y)):
        if i == 0:
            smoothed.append(y[i])
        else:
            smoothed.append(smoothed[-1] * factor + y[i] * (1 - factor))
    return np.array(smoothed)
```

```python
import matplotlib.pyplot as plt

# Smoothing factor can be adjusted (0.8 to 0.9 works well)
smoothing_factor = 0.9

# Smoothing the accuracy values
smoothed_train_accuracy = smooth_curve(emotion_model_info.history['accuracy'], factor=smoothing_factor)
smoothed_val_accuracy = smooth_curve(emotion_model_info.history['val_accuracy'], factor=smoothing_factor)

# Plotting Smoothed Training and Validation Accuracy
plt.figure(figsize=(12, 6))
plt.plot(smoothed_train_accuracy, label=' Training Accuracy', color='blue')
plt.plot(smoothed_val_accuracy, label=' Validation Accuracy', color='orange')
plt.title(' Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()
```

```python
# Smoothing the loss values
smoothed_train_loss = smooth_curve(emotion_model_info.history['loss'], factor=smoothing_factor)
smoothed_val_loss = smooth_curve(emotion_model_info.history['val_loss'], factor=smoothing_factor)

# Plotting Smoothed Training and Validation Loss
plt.figure(figsize=(12, 6))
plt.plot(smoothed_train_loss, label=' Training Loss', color='blue')
plt.plot(smoothed_val_loss, label=' Validation Loss', color='orange')
plt.title(' Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Assuming y_true and y_pred are already defined as before
Y_pred = emotion_model.predict(val_generator)
y_pred = np.argmax(Y_pred, axis=1)  # Get the predicted classes
y_true = val_generator.classes  # True labels from the generator

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)

# Emotion classes
class_names = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

# Calculate accuracy for each emotion
accuracies = {}
for i in range(len(class_names)):
    true_positives = conf_matrix[i][i]  # Diagonal elements are true positives
    total_actual = np.sum(conf_matrix[i])  # Sum of the entire row for actual emotions
    accuracy = true_positives / total_actual if total_actual > 0 else 0
    accuracies[class_names[i]] = accuracy

# Convert accuracies to a list for plotting
emotion_labels = list(accuracies.keys())
accuracy_values = list(accuracies.values())

# Plotting the accuracy for each emotion
plt.figure(figsize=(10, 6))
sns.barplot(x=emotion_labels, y=accuracy_values, palette='viridis')
plt.title('Accuracy for Individual Emotions')
plt.xlabel('Emotions')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set the y-axis limit from 0 to 1
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.grid(axis='y')  # Add grid for better visualization
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf

train_dir = '/kaggle/input/fer2013-mine/data/train'
val_dir = '/kaggle/input/fer2013-mine/data/test'

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(48,48),
    batch_size=64,  # Decreased batch size for more frequent updates
    color_mode="grayscale",
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(48,48),
    batch_size=64,  # Decreased batch size for validation consistency
    color_mode="grayscale",
    class_mode='categorical'
)

# Reduced learning rate to enable finer adjustments
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.0005,  # Slightly reduced learning rate
    decay_steps=100000,  # Decay the learning rate slower
    decay_rate=0.96,  # Gradually reduce learning rate
    staircase=True)

emotion_model = Sequential()
emotion_model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(48,48,1)))
emotion_model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2,2)))
emotion_model.add(Dropout(0.3))  # Reduced dropout for better learning capacity

emotion_model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2,2)))
emotion_model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2,2)))
emotion_model.add(Dropout(0.35))  # Balanced dropout for better generalization

emotion_model.add(Flatten())
emotion_model.add(Dense(1024, activation='relu'))
emotion_model.add(Dropout(0.5))  # Kept a high dropout for the dense layer to prevent overfitting
emotion_model.add(Dense(7, activation='softmax'))

# Use a more gradual learning rate adjustment to avoid skipping minima
emotion_model.compile(loss='categorical_crossentropy',
                      optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule),
                      metrics=['accuracy'])

print(emotion_model.summary())

emotion_model_info = emotion_model.fit(
    train_generator,
    steps_per_epoch=28709 // 64,  # Adjusted for the new batch size
    epochs=125,  # Reduced epochs since overfitting might occur at 150 epochs
    validation_data=val_generator,
    validation_steps=7178 // 64  # Adjusted for the new batch size
)

emotion_model.save_weights('model_finetuned.weights.h5')
```

```python


# Extract accuracies and validation accuracies
acc = emotion_model_info.history['accuracy']
val_acc = emotion_model_info.history['val_accuracy']

# Get alternate accuracies (1, 3, 5, etc.)
alternate_epochs = range(1, len(acc) + 1, 2)  # Start from 1, step 2
alternate_acc = acc[::2]  # Slicing every second value
alternate_val_acc = val_acc[::2]  # Slicing every second value for validation accuracy

# Plot the accuracies
plt.plot(alternate_epochs, alternate_acc, label='Training Accuracy')
plt.plot(alternate_epochs, alternate_val_acc, label='Validation Accuracy')

# Adding title and labels
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()

# Show the plot
plt.show()

```
